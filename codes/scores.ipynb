# Scores

```python
import pandas as pd
from google.colab import files

metrics = ['bertscore', 'stsb', 'meteor', 'bleu', 'gleu', 'rouge']
models = ['chatgpt1', 'chatgpt2', 'bing1', 'bing2', 'bard1', 'bard2']

name = 'chatgpt1'
df = pd.DataFrame()

for name in models:
  for metric in metrics:
    if metric == 'stsb':
      link = f'stsb_scores.csv'
      df_tmp = pd.read_csv(link)
      df['stsb'] = df_tmp[f'stsb_{name}']
    else:
      link = f'/final/scores/{name}/{name}_{metric}_scores.csv'
      df_tmp = pd.read_csv(link)

    if metric == 'bertscore':
      df = df_tmp
    elif metric == 'meteor':
      df['meteor'] = df_tmp['meteor']
    elif metric == 'bleu':
      df['bleu'] = df_tmp['bleu']
      df['brevity_penalty'] = df_tmp['brevity_penalty']
    elif metric == 'gleu':
      df['google_bleu'] = df_tmp['google_bleu']
    elif metric == 'rouge':
      df['rouge1'] = df_tmp['rouge1']
      df['rouge2'] = df_tmp['rouge2']
      df['rougeL'] = df_tmp['rougeL']

  # df.to_csv(f'{name}_scores.csv', index=False)
  # files.download(f'{name}_scores.csv')
```

```python
metrics = ['bertscore', 'stsb', 'meteor', 'bleu', 'gleu', 'rouge']
models = ['chatgpt1', 'chatgpt2', 'bing1', 'bing2', 'bard1', 'bard2']

name = 'chatgpt1'
df = pd.DataFrame()

for name in models:
  link = f'/final/scores/{name}_scores.csv'
  df_tmp = pd.read_csv(link)

  for metric in metrics:

    if metric == 'bertscore':
      df[f'{name}_{metric}'] = df_tmp['bertscore_f1']

    elif metric == 'stsb':
      df[f'{name}_{metric}'] = df_tmp['stsb']

    elif metric == 'meteor':
      df[f'{name}_{metric}'] = df_tmp['meteor']

    elif metric == 'bleu':
      df[f'{name}_{metric}'] = df_tmp['bleu']

    elif metric == 'gleu':
      df[f'{name}_{metric}'] = df_tmp['google_bleu']

    elif metric == 'rouge':
      df[f'{name}_{metric}1'] = df_tmp['rouge1']
      df[f'{name}_{metric}2'] = df_tmp['rouge2']
      df[f'{name}_{metric}L'] = df_tmp['rougeL']

df.to_csv('all_scores.csv', index=False)
files.download('all_scores.csv')
```
