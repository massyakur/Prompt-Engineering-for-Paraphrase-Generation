```python
# import the dataset from gsheets
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('5000_dataset').sheet1
```

```python
# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
# print(rows)

import pandas as pd
df = pd.DataFrame.from_records(rows)

# creating columns name
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
import time

for i in range(0, len(df), 10):
  if df.loc[i+1, "chatgpt"] != "":
    continue

  batch = df[i:i+10]
  prompt = "Paraphrase:\n"

  for _, item in batch["reference"].items():
    prompt += f"\n\"{item}\""

  for data in chatbot.ask(
        prompt,
        # auto_continue = True,
      ):
          response = data["message"]
  # print("\r", end=str(response))

  df.loc[i+1, "chatgpt"] = response
  print(i+1)
  print(response)

  time.sleep(5)

  # break
```
